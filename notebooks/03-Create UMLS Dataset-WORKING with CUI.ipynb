## MRCONSO.RRFF - `2022AB`

#### **`MRCONSO.RRF`** contains enitities from UMLS and example of this file shows as follows with information about columns:



```|C0000005|ENG|P|L0000005|PF|S0007492|Y|A26634265||M0019694|D012711|MSH|PEP|D012711|(131)I-Macroaggregated Albumin|0|N|256|```

* CUI: Concept identifier (CUI) [1] :‌ **C0000005**
* LAT: Language of Term(s) [2]: **ENG**
* TS : Term Status:‌ can be `[P, S, p, s]` [2]:‌ **P**
* LUI: Lexical identifier (LUI) [1]: **L0000005**
* STT: String Type [2]: can be `[PF, VCW, VC, VO, VW]`: **PF**
* SUI:‌ String identifier (SUI) [1] : **S0007492**
* ISPREF: Indicates whether AUI is preferred: **Y**
* AUI:‌ Atom identifier (AUI) [1]: **A26634265**
* SAUI: Source asserted atom identifier:  **""**
* SCUI: Source asserted concept identifier: **M0019694***
* SDUI: Source asserted descriptor identifier: **D012711**
* SAB: Source abbreviation: Source vocabulary [1]. [3], The source that asserts the mapping information: **MSH**
* TTY   (Term Type in Source) [2]:  **PEP**
* CODE: Unique Identifier or code for string in source: If an appropriate identifier for the map set is available from the source, it will be used as the CODE. SAUI, SCUI and SDUI may also be populated. If no source-asserted identifier is available, a CODE beginning with “MTHU” will be generated during Metathesaurus production.: **D012711**
* STR: The atom name is created as `<VSAB> to <VSAB> Mappings <optional additional information>`: **(131)I-Macroaggregated Albumin**
* SRL: Source Restriction Level and can be `[0, 1, 2, 3, 4, 9]` [2]: **0**
* SUPPRESS: Suppressible flag, can be `[E, N, O, Y]` [2]: **N**
* CVF:‌ Content view flag, **256**

Complete description of collumns in MRCONSO.RFF is avaliable here [7]. For entity name we need this file. At this file `TTY` column is an abbreviation for `term type` in source vocabulary, for example PN (Metathesaurus Preferred Name) or CD (Clinical Drug). Possible values are listed on the Abbreviations Used in Data Elements page. [9]. **THIS is not something that we should work with it for entity type task. because TTY types can be e.g:‌ AA (Attribute type abbreviation), AC‌(Activities), ..., a full list is avaliable [9]**

#### **`MRREL.RRF`** contains information about relationships in UMLS. This file is similar to the following:
|MAPSETCUI|MAPSETSAB|MAPID|MAPSID|FROMEXPR|FROMTYPE|REL|RELA|TOEXPR|TOTYPE|CVF|
|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
|C1306694|MTH|AT102971857||C0264643|CUI|SY||<Hypertension, Renovascular> AND <Hypertension, Malignant>|BOOLEAN_EXPRESSION_STR||
|C1306694|MTH|AT102971858||C0276253|CUI|SY||<Pneumonia> AND <Cytomegalovirus Infections>|BOOLEAN_EXPRESSION_STR||
|C1306694|MTH|AT102971859||C0409780|CUI|SY||<Synovitis> AND <Hand>|BOOLEAN_EXPRESSION_STR||
    
More information about columns and data structure can be found in [1] and [6].

#### **`MRSTY.RRF` file contain Semantic Types with Semantic Type tree number** [8]. In `MRSTY.RRF` each entity levels described with their valid values that has been defined in the Semantic Network. an example of this file shown as following:
|‌CUI  |TUI‌ | STN‌ | STY |

|C1898793|T204|A1.1.3|Eukaryote|AT121533921|256|
|C1898794|T005|A1.1.4|Virus|AT72943219|256|


Columns descriptions:
- CUI: Unique identifier of concept
- TUI: Unique identifier of Semantic Type
- STN: Semantic Type tree number
- STY: Semantic Type. The valid values are defined in the Semantic Network.
- ATUI: Unique identifier for attribute
- CVF:‌Content View Flag. Bit field used to flag rows included in Content View. This field is a varchar field to maximize the number of bits available for use.

-------------------------------

#### The approach for creating dataset is:
1. **Get dataset for relations detection using `MRREL.RRF` file.**
2. **Get names for CUI's using `MRCONSO.RFF` file.**
3. **Get entity types for CUI's using `MRSTY.RFF` file.**
4. **Get hierarchy of entity types using `MRSTY.RFF` file.**

````python
from tqdm.notebook import tqdm
import pandas as pd
from collections import defaultdict
from pprint import pprint
import pandas as pd
import json

def load_print_columns(path, columns_name, not_required_cols, demo):
    df_data_dict = []
    false_line = 0
    with open(path, 'r') as file_obj:
        for index_line, line in tqdm(enumerate(file_obj)):
            requireds, line_status = {}, True
            
            for index_column, line_item in enumerate(line.split("|")[:-1]):
                if columns_name[index_column] not in not_required_cols:
                    requireds[columns_name[index_column]] = line_item
                    if line_item == "":
                        line_status = False
            if line_status:    
                if demo:
                    df_data_dict.append(list(requireds.values()))
                    if index_line == 10:
                        df = pd.DataFrame(df_data_dict, columns=[column_name 
                                                                 for column_name in columns_name 
                                                                 if column_name not in not_required_cols])
                        display(df.head(4))
                        return df

                else:
                    yield requireds
            else:
                false_line += 1
    print("False lines:", false_line)

def get_generator(data):
    dataset = load_print_columns(mr_data[data]['path'], 
                             mr_data[data]['cols'], 
                             mr_data[data]['not-required-cols'], 
                             demo=False)
    return dataset

import matplotlib.pyplot as plt


def make_bar_plot(X, Y, figsize, title):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0,0,1,1])
    ax.bar(X, Y)
    ax.set_title(title)
    plt.xticks(rotation=90)
    plt.show()
````

## 5. Creating initial UMLS dataset for entity typing and relationship extraction model

````python
print("Getting SAB and STRs from MRCONSO-ENG...")
sab_names = {}
for conso in tqdm(get_generator(data = 'MRCONSO'.lower())):
    if conso['LAT'] == 'ENG':
        sab_names[conso['CUI']] = f"{conso['SAB']}@{conso['STR']}"

print("Getting CONSO-REL rels and ents based on intersections....")
conso_rel_intersection_rels_ent = defaultdict(int)
dataset = get_generator(data = 'MRREL'.lower())
sep = "\t"
for sample in tqdm(dataset):
    if sample['CUI1'] == sample['CUI2']:
        continue
    if conso_rel_intersection.get(sample['CUI1'], "NA") != "NA" and conso_rel_intersection.get(sample['CUI2'], "NA") != "NA":
        found = {}
        head, tail = False, False
        if sab_names.get(sample['CUI1'], "NA") != "NA":
            found['head'] = f"{sep}".join(sab_names[sample['CUI1']].split("@"))
            head = True
        if sab_names.get(sample['CUI2'], "NA") != "NA":
            found['tail'] = f"{sep}".join(sab_names[sample['CUI2']].split("@"))
            tail = True 
        if head and tail:
            # file.write(f"{sample['CUI1']}{sep}{sample['CUI2']}{sep}{sample['RELA']}{sep}{found['head']}{sep}{found['tail']}\n")
            conso_rel_intersection_rels_ent[f"{sample['CUI1']}{sep}{sample['CUI2']}{sep}{sample['RELA']}{sep}{found['head']}{sep}{found['tail']}"] += 1

print(f"Number of unique samples that avaliable in both MRCONSO-ENG and MRREL:{len(conso_rel_intersection_rels_ent)}")
print(f"Number of apperances of samples in both MRCONSO-ENG and MRREL:{sum(conso_rel_intersection_rels_ent.values())}")

with open(root_umls_path+"UMLS.tsv", "a") as file:
    for line in tqdm(conso_rel_intersection_rels_ent.keys()):
        file.write(str(line)+"\n")
````

````python
df = pd.read_csv(root_umls_path+"UMLS_skiped_bad_lines.tsv")[["CUI1", "CUI2", "STR-CUI1", "STR-CUI2"]]

unique_cuis = {}

for CUI, STR in zip(df['CUI1'], df['STR-CUI1']):
    if unique_cuis.get(CUI, "NA") == "NA":
        unique_cuis[CUI] = STR
print(f"Unique number of CUIs (in processing Head):", len(unique_cuis))

for CUI, STR in zip(df['CUI2'], df['STR-CUI2']):
    if unique_cuis.get(CUI, "NA") == "NA":
        unique_cuis[CUI] = STR
print(f"Unique number of CUIs (in processing Tail -- overall):", len(unique_cuis))

print("A few samples from unique_cuis:", list(unique_cuis.keys())[:20])

print(f"\nExamples: C0000052: {unique_cuis['C0000052']} , C0000172:{unique_cuis['C0000172']}")
````

````python
# get TUI from MRSTY.RFF for Entity type detections

entity_type_dict = {}

for mrsty in tqdm(get_generator(data = 'MRSTY'.lower())):
    if unique_cuis.get(mrsty['CUI'], "NA") != "NA":
        if mrsty['CUI'] in entity_type_dict.get(mrsty['CUI'], "NA") != "NA":
            entity_type_dict[mrsty['CUI']][2].append(mrsty['TUI'])
            entity_type_dict[mrsty['CUI']][3].append(mrsty['STN'])
        else:
            entity_type_dict[mrsty['CUI']] = [mrsty['CUI'], unique_cuis[mrsty['CUI']], [mrsty['TUI']], [mrsty['STN']]]

print(f"Out of {len(unique_cuis)} Unique CUIs we found Entity types for {len(entity_type_dict)}'s")
````

## 10. Create entity hierarchy

````python
stn_lst = defaultdict(int)

for sample in tqdm(get_generator(data = 'MRSTY'.lower())):
    stn_lst[sample['STN']] += 1
    
print("Number of STNs:", len(stn_lst))

make_bar_plot([str(item) for item, value in stn_lst.items()], 
              [value for item, value in stn_lst.items()], 
              figsize=(15, 3),
              title=f"Frequency plot of STNs in MRSTY (Total number of STNs:‌ {len(stn_lst)})")
````

````python
# creating Level-1
levels1 = {}
for stn in stn_lst.keys():
    if stn.split(".")[0][0] not in levels1:
        levels1[stn.split(".")[0][0]] = {stn.split(".")[0]: [stn]}
    else:
        if stn.split(".")[0] not in levels1[stn.split(".")[0][0]]:
            levels1[stn.split(".")[0][0]][stn.split(".")[0]] = [stn]
        else:
            levels1[stn.split(".")[0][0]][stn.split(".")[0]].append(stn)

for level, items in levels1.items():
    print(f"level {level} consist of {list(items.keys())} items")
    
level1s_mapper = {}
for key, value in levels1.items():
    level1s_mapper[key] = [key]
    for inner_key, inner_values in value.items():
        level1s_mapper[inner_key] = []
        for inner_value in inner_values:
            level1s_mapper[inner_key].append(inner_value)
            
print("levels1 classes are:", list(level1s_mapper.keys()))

assert len([value for _, values in level1s_mapper.items() for value in values]) == 127
````

````python
def get_level2s(l1_mapper, pre_level, level_sep_idx):
    levels = {}
    for stn in l1_mapper[pre_level]: 
        if '.'.join(stn.split(".")[:level_sep_idx]) not in levels:
            levels['.'.join(stn.split(".")[:level_sep_idx])] = [stn]
        else:
            levels['.'.join(stn.split(".")[:level_sep_idx])].append(stn)
    return levels

# deep to reach Level-2
level2s_mapper,level_sep_idx = {}, 2
for pre_level in level1s_mapper.keys(): # 'A', ['A']
                                        # 'A1', ['A1.4.1.2.1.7',...]
    if len(level1s_mapper[pre_level]) == 1:
        level2s_mapper[pre_level] = {level1s_mapper[pre_level][0]:[pre_level]}
    else:
        level2s_mapper[pre_level] = get_level2s(level1s_mapper, pre_level, level_sep_idx=level_sep_idx) 

# to make sure levels set properly
assert len([value for _, value1s in level2s_mapper.items() # 'A1', {'A1.4': ['A1.4.1.2.1.7','A1.4.1.1.1']
                  for _, values in value1s.items()         # 'A1.4', ['A1.4.1.2.1.7','A1.4.1.1.1']
                  for value in values ]) == 127
````

````python
def get_level3s(l1_mapper, pre_level, level_sep_idx):
    levels = {}
    for stn in l1_mapper:
        if '.'.join(stn.split(".")[:level_sep_idx]) not in levels:
            levels['.'.join(stn.split(".")[:level_sep_idx])] = [stn]
        else:
            levels['.'.join(stn.split(".")[:level_sep_idx])].append(stn)
    return levels

# deep to reach Level-3
level3s_mapper, level_sep_idx = {}, 3
for key1, level2_mapper in level2s_mapper.items():  # 'A', {'A': ['A']}
                                                    # 'A1' {'A1.3': ['A1.3', 'A1.3.1', 'A1.3.2', 'A1.3.3', 'A1.3.1.1']}
    level3s_mapper[key1] = {}
    for pre_level, values in level2_mapper.items(): # 'A', ['A']
                                                   # 'A1.3', ['A1.3', 'A1.3.1', 'A1.3.2', 'A1.3.3', 'A1.3.1.1']
        level3s_mapper[key1][pre_level] = {}
        if len(values) == 1:
            level3s_mapper[key1][pre_level] = {pre_level:values}
        else:
            level3s_mapper[key1][pre_level] = get_level3s(values, pre_level, level_sep_idx=level_sep_idx) 

# to make sure levels set properly            
assert len([value for _, value2s in level3s_mapper.items() # 'A1', {'A1.4': {'A1.4.1': ['A1.4.1.2.1.7']}}
                  for _, value1s in value2s.items()        # 'A1.4', {'A1.4.1': ['A1.4.1.2.1.7']}
                  for _, values in value1s.items()         # 'A1.4.1', ['A1.4.1.2.1.7']
                  for value in values]) == 127
````

````python
# Store Levels

with open(root_umls_path+"UMLS_STN_Hierarchy_level1.json", "w") as js_file:
    js_file.write(json.dumps(level1s_mapper, indent=4))

level2s = {}
for key1, value1s in level2s_mapper.items(): # 'A1', {'A1.4': ['A1.4.1.2.1.7','A1.4.1.1.1'], ...}
                                             # 'A',  {'A': ['A']}
    for key2, value2s in value1s.items():    # 'A1.4', ['A1.4.1.2.1.7','A1.4.1.1.1']
                                             # 'A', ['A']}
        if len(value2s) == 1 and value2s[0] == key1:
            continue
        level2s[key2] = []
        for value2 in value2s:              # ['A1.4.1.2.1.7','A1.4.1.1.1']
                                            # ['A']
                level2s[key2].append(value2)

with open(root_umls_path+"UMLS_STN_Hierarchy_level2.json", "w") as js_file:
    js_file.write(json.dumps(level2s, indent=4))

level3s = {}
for key2, value2s in level3s_mapper.items(): # 'A1', {'A1.4': {'A1.4.1': ['A1.4.1.2.1.7']}}
                                             # 'A',  {'A': {'A'; ['A']}}
    for key1, value1s in value2s.items():    # 'A1.4', {'A1.4.1': ['A1.4.1.2.1.7']}
                                             # 'A',  {'A': ['A']}}
            for key, values in value1s.items():         # 'A1.4.1', ['A1.4.1.2.1.7']
                                                       # 'A', ['A']}
                if len(values) == 1 and values[0] == key1:
                    continue
                level3s[key] = []
                for value in values:              # ['A1.4.1.2.1.7','A1.4.1.1.1']
                    level3s[key].append(value)

with open(root_umls_path+"UMLS_STN_Hierarchy_level3.json", "w") as js_file:
    js_file.write(json.dumps(level3s, indent=4))
````

````python
def build_hierarchy_mapper(path):
    with open(path, "r") as js_file:
        hierarchy = json.load(js_file)
    hierarchy_mapper = {item: key for key, items in hierarchy.items() for item in items}
    return hierarchy_mapper
    
entity_df = pd.read_csv(root_umls_path+"UMLS_entity_types.tsv")
stns_lst = [eval(items) for items in tqdm(entity_df['STNs'].tolist())]
print("Printing a few samples from STNs:", stns_lst[:3])

display(entity_df.head(3))

level1_mapper = build_hierarchy_mapper(root_umls_path+"UMLS_STN_Hierarchy_level1.json")
print(f"Number of STNs in level1:‌{len(level1_mapper)}")

level2_mapper = build_hierarchy_mapper(root_umls_path+"UMLS_STN_Hierarchy_level2.json")
print(f"Number of STNs in level2:‌{len(level2_mapper)}")

level3_mapper = build_hierarchy_mapper(root_umls_path+"UMLS_STN_Hierarchy_level3.json")
print(f"Number of STNs in level3:‌{len(level3_mapper)}")
````
